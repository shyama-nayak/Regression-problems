```python
def inspect_data(data):
    return pd.DataFrame({'Data Type' : data.dtypes, 'No. of levels': data.apply(lambda x: x.unique().shape[0], axis=0), 'Labels': data.apply(lambda x: str(x.unique()), axis=0)})

inspect_data(customer_data)
```

```python
customer_data.drop('CustomerID', axis=1, inplace=True)
```

```python
customer_data['City']= customer_data['City'].astype('category')
```

```python
customer_data=customer_data.loc[customer_data['MaxAgeOfChild']!=113,:]
customer_data=customer_data.loc[customer_data['MinAgeOfChild']!=0 ,:]
```

```python
y=customer_data['TotalRevenueGenerated']
X=customer_data[customer_data.columns.difference(['TotalRevenueGenerated'])]
```

```python
print(X.shape )
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3 )
print(X_train.shape)
print(X_test.shape)
```

```python
lm_formula='TotalRevenueGenerated ~ '+ '+'.join(X.columns)
print(lm_formula)
```

```python
X_train= pd.get_dummies(X_train)
```

```python
from sklearn.linear_model import LinearRegression
linearModel=LinearRegression()
linearModel.fit(X_train,y_train)
```

```python
print(linearModel.coef_)
```

```python
print(linearModel.intercept_)
```

```python
y_pred=linearModel.predict(pd.get_dummies( X_test))
```

```python
y_pred
```

```python
df=pd.DataFrame({'Actual': y_test, 'Pred': y_pred})
```

```python
df
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(mean_absolute_error(y_test, y_pred))
```

```python
print(linearModel.score(X_train, y_train))
print(linearModel.score(pd.get_dummies( X_test), y_test))
```
